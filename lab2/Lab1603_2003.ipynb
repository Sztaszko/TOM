````python
import numpy as np
import matplotlib.pylab as plt
from skimage import color
from scipy import signal
%matplotlib inline
````

Wczytanie obrazu, który mogą Państwo ściągnąć z UPEL. Oczywiście, proszę wypróbować inne obrazy. Do obsługi plików można wykorzystać:

```im.shape``` zwraca informację o rozmiarze obrazu, co można wykorzystać do warunkowego wywołania funkcji ```rgb2gray``` konwertującej do obrazu w skali szarości. W ramach teringu proszę przerobić skrypt, tak aby sprawdzać czy obraz jest w skali szarości czy też nie. Jeżeli jest kolorowy to wtedy w konstrukcji warunkowej ```if``` wywołać rgb2gray 

````python
im = plt.imread('Xray-300x247.jpg')
im.shape
im_gray = color.rgb2gray(im)
````

Przykład wyświetlenia obrazów w różnych oknach. Proszę wypróbować inne palety do wyświetlania: np. ```'hot','viridis', 'plasma', 'inferno', 'magma', 'cividis', 'jet','gray'```
Pełny wykaz palet można znaleźć na:
https://matplotlib.org/3.2.0/tutorials/colors/colormaps.html

````python
plt.figure()
plt.subplot(1, 3, 1)
plt.imshow(im)
plt.axis('off')
plt.subplot(1, 3, 2)
plt.imshow(im_gray, cmap='gray')
plt.axis('off')
plt.subplot(1, 3, 3)
plt.imshow(im_gray, cmap='jet')
plt.axis('off')
````

Przykład filtracji dowolną maską. Proszę poeksperymentować i zdefiniować swoje maski ;)

````python
#mask = np.ones((11, 11))
mask = np.array([[1,2,1],[0,0,0],[-1,-2,-1]])
display(mask)
im_filtered = signal.convolve2d(im_gray, mask)
````

I wyświetlamy:

````python
plt.figure()
plt.subplot(1, 2, 1)
plt.imshow(im_gray, cmap='gray')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(im_filtered, cmap='gray')
plt.axis('off')
````
